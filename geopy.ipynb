```python
from geopy.geocoders import Nominatim
```

```python
nom = Nominatim(user_agent='amadoune',timeout=10)
```

```python
n=nom.geocode("Paris", country_codes="fr")
print (n.latitude, n.longitude)
```

```python
#df1['postal_code']=df1.astype({'postal_code':'str'})
df1["postal_code"]=df1["postal_code"].values.astype('str')
print(df1.dtypes)
df1
```

```python
df1["address"]=df1["address"]+", "+df1["city"]+", "+df1["postal_code"]+", france"
df1
```

```python
#n=nom.geocode("Paris", country_codes="fr")
#print (n.latitude, n.longitude)

df1["coordinates"]=df1["address"].apply(nom.geocode)
df1
```

```python
df1["latitude"]=df1["coordinates"].apply(lambda x: x.latitude if x !=None else None)
```

```python
df1["Longitude"]=df1["coordinates"].apply(lambda x: x.longitude if x !=None else None)
```
